```python
import random

intentos = 1_000_000 # Para 6 decimales

def calcular(intentos):
    dentro = 0
    for _ in range(intentos):
        x = random.random()  # número entre 0 y 1
        y = random.random()

        if x**2 + y**2 <= 1:
            dentro += 1
    # Estimación de pi
    pi_estimado = 4 * dentro / intentos
    return pi_estimado

tolerancia = 0.000001  # 6 decimales
pi_real = 3.141592

pi_try = calcular(intentos)
iteraciones = 1

while abs(pi_try - pi_real) > tolerancia:
    pi_try = calcular(intentos)
    iteraciones += 1

print(f"π estimado: {pi_try}")
print(f"Iteraciones necesarias: {iteraciones}")
```
