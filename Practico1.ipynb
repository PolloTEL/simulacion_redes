 A partir de acá abajo cambié un poco el esquema del algoritmo, (los parametros de término y los intentos acumulados) Esto debido a que se estaba tardando más de 5 minutos.

```python
import random
import math
import time

tiempo_inicio = time.time()
tolerancia = 0.000001 # Para 6 décimales

intentos = 0
dentro = 0
pi_try = 0

while abs(pi_try - math.pi) > tolerancia or intentos == 0:
    x = random.random()  # número entre 0 y 1
    y = random.random()

    if x**2 + y**2 <= 1:
        dentro += 1

    intentos += 1
    pi_try = 4 * dentro / intentos

tiempo_final = time.time()

print(f"Pi calculado: {pi_try:.6f}.")
print(f"Pi real: {math.pi:.6f}.")
print(f"Iteraciones totales: {iteraciones}.")
print(f"Tiempo total: {(tiempo_final - tiempo_inicio):.2f} segundos.")
```
